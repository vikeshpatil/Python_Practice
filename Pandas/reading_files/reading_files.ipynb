# Indexing and Slicing Dataframe Tables

##### We can access the data by slicing. Following are the three methods useds for slicing data from above table

##### 1) .loc[starting index: last index, Starting column: endcolumn]
        In this method we use the lables of columns and indices


```python
df5.loc["735 Dolores St":"1056 Sanchez St", "State": "Name"]
```

##### 2) .iloc[ , ]
        In this method we use the actual index of the coloumns as well


```python
df5.iloc[1:4, 2:5]
```

##### 3) .ix[ , ]
        In this method, we can use both index and lables to access data.
        

```python
df5.ix[2:5, "State":"Name"]
```

## Deleting columns and rows

        using drop("Column/Index name", 1/0) 
            If it is column then use 1 otherwise if it is a row then use 0

```python
df5
```

```python
df5.drop("City", 1)   #Deletes City Column
```

```python
df5.drop("332 Hill St", 0)  #Deletes 332 Hill St row
```

```python
df5.drop(df5.columns[0:3], 1)    #It will delete multiple columns supplied to method
```

```python
df5.drop(df5.index[1:4])     #It will delete multiple rows with supplied indices
```
